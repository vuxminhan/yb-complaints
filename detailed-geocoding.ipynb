```python
geocoded = pd.read_csv('complaints-all-reversed-geocoded.csv')
geocoded.head()
```

```python
import ast

geocoded['Address'] = geocoded['Address'].apply(ast.literal_eval)
geocoded['Address']
```

```python
def get_attr(x, attr):
    # Safely check if the 'district' in 'compound' is None
    if x.get('compound') is None or x.get('compound', {}).get('district') is None:
        # Split the formatted_address into a list based on commas
        address_parts = x['formatted_address'].split(',')
        address_parts = [part.strip() for part in address_parts]  # Remove any leading/trailing whitespace

        # Handle different attributes based on their expected positions
        if attr == 'province' and len(address_parts) >= 1:
            return address_parts[-1]
        elif attr == 'commune' and len(address_parts) >= 2:
            return address_parts[-2]
        elif attr == 'district' and len(address_parts) >= 3:
            return address_parts[-3]
        else:
            return None
    # elif x['compound']['province'] is None:

    else:
        try:
            return x['compound'][attr]
        except:
            return None

geocoded['District'] = geocoded['Address'].apply(lambda x: get_attr(x, 'district'))
geocoded['Commune'] = geocoded['Address'].apply(lambda x: get_attr(x, 'commune'))
geocoded['Province'] = geocoded['Address'].apply(lambda x: get_attr(x, 'province'))

geocoded
```

```python
geocoded.to_csv('complaints-all-reversed-geocoded-granulated.csv', index=False)
```
