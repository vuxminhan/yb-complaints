```python
geocoded = pd.read_excel("complaints-all-geocoded-granulated.xlsx")
```

```python
geojson = {}
features = []
for i, r in geocoded.iterrows():
    entry = {
        "type": "Feature",
        "properties": {
            "field": r['field'],
            "subfield": r['subfield'],
            "sentiment": r['sentiment'],
            "district": r['District'],
            "commune": r['Commune'],
            "province": r['Province']
        },
    }
    entry["geometry"] = {
        "type": "Point",
        "coordinates": [r['Coordinates']['lng'], r['Coordinates']['lat']]
    }
    features.append(entry)
geojson["type"] = "FeatureCollection"
geojson["crs"] = {
        "type": "name",
        "properties": {
            "name": "urn:ogc:def:crs:OGC:1.3:CRS84"
        }
    }
geojson["features"] = features
```

```python
import json

# Assuming 'geojson' is your data structure containing the GeoJSON data
with open('complaints.geojson', 'w', encoding='utf-8') as file:
    json.dump(geojson, file, ensure_ascii=False)
```
